[HYG Database](http://www.astronexus.com/hyg)
contains all ~120,000 stars that are either:  

- brighter than a certain magnitude cutoff (magnitude +7.5 to +9.0) or 
- within 50 parsecs (about 160 light years) from the Sun.""


```python
import pandas as pd
import numpy as np
df = pd.read_csv('http://www.astronexus.com/files/downloads/hygdata_v3.csv.gz')
```

HYG dataset reports flux ratio 'color index' $B-V$; can be transformed to temperature using:
$${\displaystyle T=4600\,\mathrm {K} \left({\frac {1}{0.92(B-V)+1.7}}+{\frac {1}{0.92(B-V)+0.62}}\right).}$$
(Ballesteros 2012)

```python
df.loc[:,'temp'] = 4600 * (1 / (0.92 * df.ci + 1.7) + 1 / (0.92 * df.ci + 0.62))
df.loc[:,'logtemp'] = np.log10(df.temp)
df.loc[:,'loglum'] = np.log10(df.lum)
```

```python
import altair as alt; 
alt.data_transformers.enable('json')
alt.renderers.enable('notebook')

data = df.sample(1000)
star_chart = alt.Chart(data).mark_circle(size=1).encode(
    x=alt.X('temp', 
            axis=alt.Axis(title='Temperature (K)')),
    y=alt.Y('lum', 
            axis=alt.Axis(title=r'Luminosity (L)'),
            scale=alt.Scale(type='log')),
    color=alt.Color('temp', scale=alt.Scale(range=['red','blue'])))
```

```python
import numpy as np
from sklearn import mixture, cluster, neighbors
n_clusters = 10

from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.cubehelix_palette(rot=2, n_colors=n_clusters))
                      
bgmm = mixture.BayesianGaussianMixture(n_components=n_clusters, 
    weight_concentration_prior_type='dirichlet_process',
    weight_concentration_prior=1e5)
```

```python
cluster_data = df.loc[:, ['logtemp', 'loglum']].dropna().sample(2000)
bgmm.fit(cluster_data)
bgmm_labels = bgmm.predict(cluster_data)
cluster_data.plot.scatter(x='logtemp', y='loglum', 
                          c=bgmm_labels, colormap=cmap,
                          s=5)
plt.gca().invert_xaxis()
```
